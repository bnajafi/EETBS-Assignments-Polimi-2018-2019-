```python
ExternalFilesFolder = "/Users/federicarossetti/Documents/python4ScientificComputing_Numpy_Pandas_MATPLotLIB/ExternalFiles"
ConsumptionFileName = "consumption_5545.csv"
TemperatureFileName = "Austin_weather_2014.csv"
IrradianceFileName = "irradiance_2014_gen.csv"
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_consumptionFile = os.path.join(ExternalFilesFolder, ConsumptionFileName)
path_Temperaturefile = os.path.join(ExternalFilesFolder, TemperatureFileName)
path_Irradiance = os.path.join(ExternalFilesFolder, IrradianceFileName)
```

CONSUMPTION:

```python
DF_consumption=pd.read_csv(path_consumptionFile,sep=",",index_col=0)
PreviousIndex=DF_consumption.index
NewParsedIndex=pd.to_datetime(PreviousIndex)
DF_consumption.index=NewParsedIndex
DF_consumption_somedaysMay=DF_consumption["2014-05-15 00:00:00":"2014-05-30 23:00:00"]
```

```python
DF_consumption_somedaysMay
```

TEMPERATURE:

```python
DF_weather=pd.read_csv(path_Temperaturefile, sep=";", index_col=0) 
PreviousIndex_weather=DF_weather.index
newIndex_weather=pd.to_datetime(PreviousIndex_weather)
DF_weather.index=newIndex_weather
DF_weather.columns
DF_Temperature=DF_weather["temperature"]
DF_Temperature_somedaysMay=DF_Temperature["2014-05-15 00:00:00":"2014-05-30 23:00:00"]
```

```python
DF_Temperature_somedaysMay
```

IRRADIANCE:

```python
DF_irradianceSource=pd.read_csv(path_Irradiance, sep=";", index_col=1)
DF_irradiance=DF_irradianceSource[["gen"]]
DF_irradiance[DF_irradiance["gen"]<0]=0
PreviousIndex_irradiance=DF_irradiance.index
newIndex_irradiance=pd.to_datetime(PreviousIndex_irradiance)
DF_irradiance.index=newIndex_irradiance
DF_irradiance_somedaysMay=DF_irradiance["2014-05-15 00:00:00":"2014-05-30 23:00:00"]
```

Plotting the graphs:

```python
fig,ax=plt.subplots(3) 
ax[0].plot(DF_consumption_somedaysMay, color="blue",label="Consumption")
ax[0].legend()
ax[1].plot(DF_Temperature_somedaysMay, color="r",label="Temperature")
ax[1].legend()
ax[2].plot(DF_irradiance_somedaysMay, color="g",label="Irradiance")
ax[2].legend()
plt.xlabel("Time")
plt.show()
```

To create a unique graph:

```python
DF_joined=DF_consumption_somedaysMay.join([DF_Temperature_somedaysMay,DF_irradiance_somedaysMay])


temp_min= DF_joined["temperature"].min() 
temp_max= DF_joined["temperature"].max() 

irrad_min= DF_joined["gen"].min() 
irrad_max= DF_joined["gen"].max()

cons_min= DF_joined["air conditioner_5545"].min() 
cons_max= DF_joined["air conditioner_5545"].max()

DF_joined["consumption normalized"]=(DF_joined["air conditioner_5545"]-cons_min)/(cons_max-cons_min)
DF_joined["temperature normalized"]=(DF_joined["temperature"]-temp_min)/(temp_max-temp_min)
DF_joined["irradiance normalized"]=(DF_joined["gen"]-irrad_min)/(irrad_max-irrad_min)
```

```python
plt.figure()
DF_joined["consumption normalized"].plot()
DF_joined["temperature normalized"].plot()
DF_joined["irradiance normalized"].plot()
plt.legend()
plt.xlabel("Time")
plt.ylabel("AC_power (W)")
plt.show()
```
